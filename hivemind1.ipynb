```
#Packages
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
%matplotlib
```

```

#URLs contaning the raw csv files
url_2015 = "https://raw.githubusercontent.com/Priyatham-sai-chand/Happiness/master/2015.csv"
url_2016 = "https://raw.githubusercontent.com/Priyatham-sai-chand/Happiness/master/2016.csv"
url_2017 = "https://raw.githubusercontent.com/Priyatham-sai-chand/Happiness/master/2017.csv"

#intialising dataframes with the URLs Content
df_2015 = pd.read_csv(url_2015)
df_2016 = pd.read_csv(url_2016)
df_2017 = pd.read_csv(url_2017)
```

# Question 1 

**Countries with a happiness score of less than 5.0**


```
def happiness_less_than_score(df,score):
  """Gathers the country names with happiness score less than specified."""
  if df.equals(df_2015):
    return df_2015['Country'][df_2015['Happiness Score'] < score]
  elif df.equals(df_2016):
    return df_2016['Country'][df_2016['Happiness Score'] < score]
  elif df.equals(df_2017):
    return df_2017['Country'][df_2017['Happiness.Score'] < score]
  

```

**Answer**

```
### Run this cell ###

result = happiness_less_than_score(df_2015,5.0)
print("In Year 2015: ")
print(result.tolist()) #pprint(result.tolist())
print()
result = happiness_less_than_score(df_2016,5.0)
print("In Year 2016: ")
print(result.tolist()) #pprint(result.tolist())
print()
result = happiness_less_than_score(df_2017,5.0)
print("In Year 2017: ")
print(result.tolist()) #pprint(result.tolist())
```

# Question 2

 **Which is the unhappiest country in Sub-Saharan Africa?**

```
def unhappiest_in_region(df,region):
  """ Finds the least happiness score in the specified region."""
  if df.equals(df_2015):
    df_temp = df_2015[df_2015['Region'] == region]
    return df_temp[df_temp['Happiness Score'] == df_temp['Happiness Score'].min()]
  elif df.equals(df_2016):
    df_temp = df_2016[df_2016['Region'] == region]
    return df_temp[df_temp['Happiness Score'] == df_temp['Happiness Score'].min()]
  elif df.equals(df_2017):
    # Since 2017 report doesn't contain regions they are explictly taken from the 2015 year's countries under that region.

    df_region_country =  df_2015['Country'][df_2015['Region'] == region].to_frame()
    df_temp = df_2017[df_2017['Country'].isin(df_region_country['Country'])]
    return df_temp[df_temp['Happiness.Score'] == df_temp['Happiness.Score'].min()]


```

**Answer:**

```
### Run this cell ###

result = unhappiest_in_region(df_2015,"Sub-Saharan Africa")
print("In Year 2015: ")
print(result['Country'].tolist()) 
print()
result = unhappiest_in_region(df_2016,"Sub-Saharan Africa")
print("In Year 2016: ")
print(result['Country'].tolist()) 
print()
result = unhappiest_in_region(df_2017,"Sub-Saharan Africa")
print("In Year 2017: ")
print(result['Country'].tolist()) 
```

```
def happiest_in_region(df,region):
  """ Finds the most happiness score in the specified region."""
  if df.equals(df_2015):
    df_temp = df_2015[df_2015['Region'] == region]
    return df_temp[df_temp['Happiness Score'] == df_temp['Happiness Score'].max()]
  elif df.equals(df_2016):
    df_temp = df_2016[df_2016['Region'] == region]
    return df_temp[df_temp['Happiness Score'] == df_temp['Happiness Score'].max()]
  elif df.equals(df_2017):
    # Since 2017 report doesn't contain regions they are explictly taken from the 2015 year's countries under that region.

    df_region_country =  df_2015['Country'][df_2015['Region'] == region].to_frame()
    df_temp = df_2017[df_2017['Country'].isin(df_region_country['Country'])]
    return df_temp[df_temp['Happiness.Score'] == df_temp['Happiness.Score'].max()]
```

**Answer**

Happiness Score

A metric measured in 2015 by asking the sampled people the question: "How would you rate your happiness on a scale of 0 to 10 where 10 is the happiest."

As per the Metric, All the countries above 5 are considered to be "happy" and all the countries below the score of 5 are considered to be "unhappy".

```
def happiness_more_than_score(df,score):
  """Gathers the country names with happiness score greater than specified."""
  
  if df.equals(df_2015):
    return df_2015['Country'][df_2015['Happiness Score'] > score]
  elif df.equals(df_2016):
    return df_2016['Country'][df_2016['Happiness Score'] > score]
  elif df.equals(df_2017):
    return df_2017['Country'][df_2017['Happiness.Score'] > score]
```

**Answer**

Importing and Downloading the files

```
import os
import requests
from tqdm.auto import tqdm
from pyunpack import Archive
from clint.textui import progress
import shutil

labels_url = "https://www.crcv.ucf.edu/data/UCF101/UCF101TrainTestSplits-RecognitionTask.zip"
data_url = "https://www.crcv.ucf.edu/data/UCF101/UCF101.rar"


os.chdir('/content')

# labels file zip download
r = requests.get(labels_url, stream=True)
file_name = labels_url.split('/')[-1]
with tqdm.wrapattr(open(file_name, "wb"), "write", miniters=1,
                   total=int(r.headers.get('content-length', 0)),
                   desc=file_name) as fout:
    for chunk in r.iter_content(chunk_size=4096):
        fout.write(chunk)

os.chdir('/content')

# data file zip download
r = requests.get(data_url, stream=True)
file_name = data_url.split('/')[-1]
with tqdm.wrapattr(open(file_name, "wb"), "write", miniters=1,
                   total=int(r.headers.get('content-length', 0)),
                   desc=file_name) as fout:
    for chunk in r.iter_content(chunk_size=4096):
        fout.write(chunk)

  
```
